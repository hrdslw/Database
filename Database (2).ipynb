```python
# importting Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings as wr
wr.filterwarnings('ignore')

# Считывание данных из файла и вывод информации о нем
#df - dataframe
df = pd.read_excel("dataset_mental_report_new.xlsx")
df.drop(["Unnamed: 0"], axis = 1, inplace = True) #Удалил первый столбец.

#Plots Creation
age_group_counts = df['old_group'].value_counts()
plt.figure(figsize = (16,9))
plt.bar(age_group_counts.index, age_group_counts, color = 'green')
plt.xlabel('Age Group')
plt.ylabel('Counts')
plt.title('Count Plot of Ages')
plt.show()




# Swarmplot для распределения по полу и возрастным группам на основе времени Stroop
plt.figure(figsize = (16, 9))
sns.swarmplot(x ='old_group', y ='stroop_duration', hue ='gender', data = df)
plt.title("Swarmplot: Время Stroop по возрастным группам и полу")
plt.show()


#Violinplot
plt.figure(figsize = (16,9))
sns.violinplot(data = df, x = 'old_group', y = 'stroop_duration', hue = 'gender')
plt.title("Violinplot: Время Stroop по возрастным группам и полу")
plt.show()

# Boxplot
plt.figure(figsize = (6, 12))
sns.boxplot(y = df['stroop_duration'])
plt.title('Боксплот для времени Stroop')
plt.show()

#Heatmap
data = df.head(10).pivot(index="old", columns="num", values="RR_dev")
plt.figure(figsize = (15, 10))
sns.heatmap(data)
plt.title("RR_dev Heatmap")
plt.show()
```
